```python
# 建立原始 ChatGLM2 模型
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm2-6b", trust_remote_code=True)
model = AutoModel.from_pretrained("THUDM/chatglm2-6b", trust_remote_code=True, device='cuda')
model = model.eval()
```

```python
# 問答測試-1

q = "请假装你是一个心理咨询师，根据下列发问者所提供之问题进行思考，并以第一人称口吻给出答复。现在的我，感觉就像是从自己出生以来就是多余的，我就不应该出现在这个世界，现在就读高三的我，面对家庭，学校，朋友，老师，我都是被否定的，我真的感觉好难过好像一切都不需要我，我对他们来说就是可有可无的，我现在心里只想高考，但是学习成绩不好常常引来怀疑，我每天真的过得很累，无朋友，只能自己一个人，我可以不要这些，但是我现在真的很迷茫，我想高考，我想考本科，可是，我不知道该从何说起，我不管再怎么努力也没有进步，也许你们会说我学习方法有问题，可从来没人和我说过应如何改正，我真的很累，我感觉现在我的生活毫无意义，我不知道该怎样去面对高考，我总是相信我努力了总会上本科，可是每次的考试都是给了我一次又一次的打击，我不知道我到底该如何学习了，我迷茫了，现在我连我最好的英语和物理都考不好，还说其他的，谢谢你的阅读，我真的想好好参加高考，我不想辜负他们的期望，看到信的您，能否给我一些建议么，谢谢您"
response, history = model.chat(tokenizer, q, history=[], top_p=0.3, do_sample=True, repetition_penalty=1.2)
print(response)
```
